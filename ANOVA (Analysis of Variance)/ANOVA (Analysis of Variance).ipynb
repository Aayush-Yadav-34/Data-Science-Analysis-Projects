### **ANOVA (F-TEST)**

**One Way F-test (Anova):** We will compare the means of **Credit Score** across the three different **Geography** groups (France, Spain, Germany) to see if they are similar.

```python
import pandas as pd
import scipy.stats as stats

# Load the dataset
df = pd.read_csv("Churn_Modelling.csv")

# Sample data: We will use the 'CreditScore' column
data = df['CreditScore']
# Prepare data for ANOVA
d_data = {grp: df['CreditScore'][df.Geography == grp] for grp in df.Geography.unique()}

# Perform One-way ANOVA
F, p = stats.f_oneway(d_data['France'], d_data['Spain'], d_data['Germany'])

print("p-value for significance is: ", p)

if p < 0.05:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```

### **Two Way F-test**

We will check the effect of two independent variables (**Geography** and **Gender**) on a dependent variable (**EstimatedSalary**).

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Fit the OLS model
# We are testing: EstimatedSalary ~ Geography + Gender + Geography*Gender
model = ols('EstimatedSalary ~ C(Geography)*C(Gender)', df).fit()

print(f"Overall model F({model.df_model: .0f},{model.df_resid: .0f}) = {model.fvalue: .3f}, p = {model.f_pvalue: .4f}")

# Perform ANOVA on the model
res = sm.stats.anova_lm(model, typ=2)
print(res)
```
